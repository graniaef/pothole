```python
# Create .yaml file 
import yaml

data_yaml = dict(
    train = '/kaggle/input/pothole-dataset/datasetfix/train',
    val = '/kaggle/input/pothole-dataset/datasetfix/val',
    nc = 3,
    names = ['crack', 'good', 'plothole'])

# Note that I am creating the file in the yolov5/data/ directory.
with open('data.yaml', 'w') as outfile:
    yaml.dump(data_yaml, outfile, default_flow_style=True)
```

```python
from ultralytics import YOLO

# Load model YOLOv8s
model = YOLO('yolov8s.pt')

# Latih model
model.train(data='data.yaml', epochs=50, imgsz=640, batch=32)
```

# Evaluasi Model

```python
import cv2
import matplotlib.pyplot as plt
import os
from ultralytics import YOLO


model = YOLO('/kaggle/working/pothole.pt')

dataset_path = '/kaggle/input/evaluasimodel/'

output_path = '/kaggle/working/'

image_files = [f for f in os.listdir(dataset_path) if f.endswith('.jpg')]  

for image_file in image_files:
    image_path = os.path.join(dataset_path, image_file)
    results = model(image_path)
    result = results[0]
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    for box in result.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())
        label = int(box.cls.item())  
        conf = box.conf.item()  
        cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 0), 2)
        cv2.putText(img, f'{label} {conf:.2f}', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

    output_image_path = os.path.join(output_path, image_file)
    cv2.imwrite(output_image_path, cv2.cvtColor(img, cv2.COLOR_RGB2BGR))
    
    plt.imshow(img)
    plt.title(image_file)  
    plt.axis('off')
    plt.show()
```
